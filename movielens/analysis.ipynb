### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times 

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# group movie_df by movie id and add all counts
movie_agg_df=rating_df.groupby("movie id").count() 

# group movie_df by movie id and add all sums
movie_agg_df["rating_sum"]=rating_df.groupby("movie id").rating.sum()
movie_agg_df
```

```python
# filter movies with 100+ ratings and stord in top_movie_agg dataframe
top_movie_agg=movie_agg_df[movie_agg_df["rating"]>=100]

# add average rating as column to dataframe
top_movie_agg["avg_rating"]=top_movie_agg["rating_sum"]*1.00/top_movie_agg["rating"]*1.00

# sort by average rating and filter by top 25 and store in a separate dataframe top25_movie_agg
top_25_movie_agg=top_movie_agg.sort_values(by="avg_rating",ascending = False).head(25)
```

```python
# prep movie_df for join and set the movie id as index to join movie id
movie_df_lookup=movie_df[["movie id","movie title"]]
movie_df_lookup.index = movie_df_lookup["movie id"]

# join Top 25 movies average rating with movie_df for the final report
top_25_report=top_25_movie_agg.join (movie_df_lookup,how="left")[["movie title", "avg_rating"]]
top_25_report.reset_index()
```

### 10. Check for the validity of the below statements with respect to the data provided

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

**Please pay attention to what should be the denominator while calculating percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#rating_df left join with user_df on user id to get gender
rating_gender_df=pd.merge(rating_df, user_df[["user id","gender"]], on="user id", how="left")


#prep the movie_df for this analysis

#add a column called total to sum all genres at row level
movie_df['total']=movie_df.iloc[:, 2:].sum(axis=1) # exclude index and first 2 columns in row sum

#add column other to sum all non (drama/sci-fi/romance) at row leve=l
movie_df["other"] = movie_df["total"]- (movie_df["Drama"]+movie_df["Sci-Fi"]+movie_df["Romance"])


#rating_gender_df left join with movie df to get all genres counts
rating_gender_movie_df=pd.merge(rating_gender_df,movie_df [['movie id','Drama','Sci-Fi','Romance','other','total']]
                                , on="movie id", how="left")

# group by gender and sum all genres
gender_groupby=rating_gender_movie_df.groupby ("gender").sum()
                                

# add percentage numbers
gender_groupby["Sci-Fi perc"]=gender_groupby["Sci-Fi"]/gender_groupby["total"]
gender_groupby["Drama perc"]=gender_groupby["Drama"]/gender_groupby["total"]
gender_groupby["Romance perc"]=gender_groupby["Romance"]/gender_groupby["total"]
gender_groupby
         
```
